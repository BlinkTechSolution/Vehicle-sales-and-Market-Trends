```python
# Function to remove duplicates from the model column
def sentence_case(text):
    if pd.isnull(text):  # Check for NaN values
        return text
    text = str(text).lower()  # Convert the text to lowercase
    return text.capitalize()  # Capitalize only the first letter
df['model'] = df['model'].apply(sentence_case)
```

```python
# Create a column for the vehicles manufacturer number
df['Vin_maker'] = df['vin'].str[:4]
```

```python
# Create a dictionary of manufacturer name with their vin numbers
data = df.groupby('make')['Vin_maker'].unique()
data.to_dict()
```

```python
# Function to replace the missing values in the make column.
def maker_vectorized(df, data):
    # Create a function that searches for the key in the dictionary
    def find_maker(vin_value):
        for key, value in data.items():
            if str(vin_value) in value:  # Search for the vin_value in the list
                return key
            
        return None  # Default value if no match is found

    # Apply the function to each row in the 'Vin_maker' column
    df['make'] = df['Vin_maker'].apply(find_maker)
    
    return df['make']
```

```python
# Application of the created function
df['make'] = maker_vectorized(df, data)
```

# Identify trends in vehicle sales over time.

```python
# Create a column for the saledate (years) 
df['sale_year'] = df['saledate'].dt.year
```

```python
# Print out sales that took place in year 2014
df1= df[df['sale_year']==2014]
df1
```

```python
# Arrange the saledate of 2014 for analysis 
df1.set_index('saledate', inplace=True)

# Resample data by month and sum the sales
monthly_sales = df1['sellingprice'].resample('M').sum()

# Check the aggregated monthly sales data
print(monthly_sales)
```

```python
# Plot graph of number of vehicle sales for year 2014
plt.figure(figsize=(10,6))
plt.plot(monthly_sales.index, monthly_sales, marker='o')
plt.title('Vehicle Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Vehicles Sold')
plt.grid(True)
plt.show()
```
